```python
import yt_dlp
import google.generativeai as genai

# üîê Configure Gemini API
genai.configure(api_key="")
model = genai.GenerativeModel("gemini-2.0-flash")
chat = model.start_chat(history=[])

# üì• Get video link
link = input("üîó Enter the YouTube video link: ")

try:
    # üì¶ Extract info using yt_dlp
    ydl_opts = {
        'quiet': True,
        'skip_download': True,
        'force_generic_extractor': False,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(link, download=False)

    title = info.get("title", "N/A")
    views = info.get("view_count", "N/A")
    duration = info.get("duration", "N/A")
    rating = info.get("average_rating", "N/A")
    description = info.get("description", "No description found.")

    print("\n--- Video Details ---")
    print("üìå Title      :", title)
    print("üëÅÔ∏è  Views     :", views)
    print("‚è±Ô∏è  Duration  :", duration, "seconds")
    print("‚≠ê Rating     :", rating)
    print("\nüìù Description:\n", description)

    # üí¨ Start interactive Q&A based on description
    print("\nüí° You can now ask questions about the video description.")
    print("Type 'exit' to stop chatting.\n")

    chat.send_message(f"You are an AI tutor. Here's a YouTube video description:\n\n{description}\n\nUse this as reference to answer upcoming questions.")

    while True:
        query = input("You: ")
        if query.lower() == "exit":
            print("üëã Exiting chat.")
            break

        response = chat.send_message(query)
        print("ü§ñ Gemini:", response.text)

except Exception as e:
    print("‚ùå An error occurred:", str(e))
```
